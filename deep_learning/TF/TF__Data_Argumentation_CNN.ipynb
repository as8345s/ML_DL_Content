<h1>Data Argumentation CNN</h1>

Worum geht es hier? <br>
Nehmen wir ein CNN Netzwerk und trainieren es mit einem Dataset das aus Blumen besteht. Das Ziel ist, Blumen mit einem CNN zu klassifizieren. Mit einem CNN geht das deutlich besser als mit einem ANN durch das Vorschalten eines Faltungsnetzes für die Extrahierung von Features.

Aber was ist, wenn nach dem Trainieren Blumenbilder kommen, die rotiert sind, oder eine andere Transformation aufweisen.
- Es kann zu Overfitting einer Blume kommen, da diese immer in der gleichen/ähnlichen Position ist.
- Um das Problem anzugehen: Data-Argumentation

```python
# Imports.
import tensorflow as tf
import numpy as np
import PIL
import cv2   
import pathlib

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Hole Dataset
# - Siehe Tensorflow: https://www.tensorflow.org/datasets/catalog/tf_flowers
url = "http://download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=url, cache_dir='./data/datasets/tf_flowers', untar=True)
```

```python
# Einfacher mit Pfaden umzugehen. 
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
# Liste Verzeichnisse auf uns zähle wie viele es sind.
len(list(data_dir.glob("*/*.jpg")))
# - Oder <Klasse>/<name>.jpg
```

```python
flowers = ['daisy', 'dandelion', 'roses', 'sunflowers', 'tulips'] 
```

PIL eignet sich sehr gut, um Bilder aus einem Verzeichnis zu laden und zu zeigen.

```python
# Alternativ erstelle pro Klasse eine Liste.
PIL.Image.open( str( list(data_dir.glob("sunflowers/*.jpg"))[0]  ) )
```

Um alle Bilder zu nutzen, kann z. B. ein Dict, Array, usw. angelegt werden, um die Abfrage zu erleichtern.

```python
list_flowers = {}
for i in flowers:
    list_flowers[i] = list(data_dir.glob(f"{i}/*.jpg"))
```

Beim Betrachten den verschiedenen Bildern kann man feststellen, dass diese sich in der Größe unterscheiden.

```python
# Mit resize Größe anpassen.
img = cv2.resize( cv2.imread(list_flowers['daisy'][4]), (200, 200))
img.shape
# Für das Training müssen alle Bilder dieselbe Größe haben. 
```

Um ein Dataset zu erstellen, kann mit einer Schleife durchiteriert werden, und auf den Weg dorthin kann auch optional eine Transformation durchgeführt werden.

```python
# Erstelle Dataset.:
X_data = [] # Blumen.
y_data = [] # Labels. 

for key, items in list_flowers.items():

    for item in items:
        img = cv2.resize( cv2.imread(item), (195, 195))
        X_data.append(img)
        y_data.append(key)
```

```python
y_data[0]
```

Dann müssen die y-Daten umgeformt werden.

```python
name_dict = {'daisy':0, 'dandelion':1, 'roses':2, 'sunflowers':3, 'tulips':4 }
y_data_new = [name_dict[item]  for item in y_data]        
```

```python
y_data_new[:3]
```

```python
# Als Numpy.
X_data = np.array(X_data)
del y_data
y_data = np.array(y_data_new)
del y_data_new
```

Jetzt kommt das Skalieren und die Aufteilung. 

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2)
```

```python
# Skaliere.
X_train_sc = X_train / 255
X_test_sc  = X_test  / 255
```

<h2>Model 1</h2>

Das erste Model wird ohne einsatz von Data-Argumentation trainiert.

```python
model = tf.keras.Sequential([
    # Faltungsnetz. 
    tf.keras.layers.Conv2D(filters=30, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(filters=40, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(filters=50, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    # ANN Dense.
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5)  
])

tb_callback = tf.keras.callbacks.TensorBoard(log_dir="logs/data_arg", histogram_freq=1)

model.compile(
    optimizer = 'adam',
    loss      = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics   = ['accuracy']
)
```

```python
model.fit(X_train_sc, y_train, epochs=7, callbacks=[tb_callback])
```

```python
# Teste Model.
model.evaluate(X_test_sc, y_test)
```

Beim Training erzielt das Model eine gute Genauigkeit, beim Testen aber viel weniger => Overfittig
- Als Lösung: Data-Agumentation

<h2>Model 2</h2>

Mit Tensorflow kann das problem einfach angegangen werden.

```python
# Transformation
# - Siehe: https://www.tensorflow.org/api_docs/python/tf/keras/layers/ <Methode> 
# - Es gibt verschiedene Methoden, um ein Bild zu verändern. 
transformation = tf.keras.Sequential([
   tf.keras.layers.RandomFlip("horizontal", input_shape=(195, 195, 3)),
   tf.keras.layers.RandomZoom(0.2),
   tf.keras.layers.RandomRotation(0.2),
])
```

```python
model2 = tf.keras.Sequential([
    # Faltungsnetz.
    transformation,
    tf.keras.layers.Conv2D(filters=30, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(filters=40, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(filters=50, padding='same', kernel_size=(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    # ANN Dense.
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5)  
])

tb_callback = tf.keras.callbacks.TensorBoard(log_dir="logs/data_arg_2", histogram_freq=1)

model2.compile(
    optimizer = 'adam',
    loss      = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics   = ['accuracy']
)
```

```python
model2.fit(X_train_sc, y_train, epochs=7, callbacks=[tb_callback])
```

```python
model2.evaluate(X_test_sc, y_test)
```

Man sieht, dass die Genauigkeit des Trainings etwas gesunken ist, dafür ist die Genauigkeit beim Testen gestiegen. Es generalisiert besser. 
- Bei mehr Layers und Einstellungen ist der Effekt noch besser sichtbar.
